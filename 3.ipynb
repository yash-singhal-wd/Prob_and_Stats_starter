```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(0)
data = np.random.normal(loc=50, scale=10, size=1000)  # Mean=50, Std=10

mean = np.mean(data)
std_dev = np.std(data)
z_data = (data - mean) / std_dev

sns.set(style="whitegrid")

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.kdeplot(data, fill=True, color='skyblue')
plt.title("Original Normal Distribution\nMean = {:.2f}, Std = {:.2f}".format(mean, std_dev))
plt.xlabel("Original Data")

plt.subplot(1, 2, 2)
sns.kdeplot(z_data, fill=True, color='salmon')
plt.title("Standard Normal Distribution (Z)\nMean ≈ {:.2f}, Std ≈ {:.2f}".format(np.mean(z_data), np.std(z_data)))
plt.xlabel("Z-Score")

plt.tight_layout()
plt.suptitle("Standard Normal Variate Transformation", y=1.05, fontsize=16)
plt.show()
```
